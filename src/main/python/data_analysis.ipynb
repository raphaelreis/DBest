```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
root = "../../../"
filePath = root + "data/store_sales_sample.dat"
df = pd.read_csv(filePath, sep="|", header=None).iloc[:, 11:13]
df.columns = ["ss_wholesale_cost", "ss_list_price"]
```

```python
df.ss_list_price.corr(df.ss_wholesale_cost)
```

```python
df.ss_list_price.to_numpy()
```

```python
gradient, intercept, r_value, p_value, std_err = stats.linregress(df.ss_list_price.to_numpy(), df.ss_wholesale_cost.to_numpy())
```
